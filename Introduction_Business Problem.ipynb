# Introduction to Business Problem

## Introduction

Imagine you are considering to invest in a business idea for a musical instruments store in Toronto. At the outset, there are numerous factors to take into consideration in order to evaluate this business idea such as determing product catalogue, delineating business structure and marketing strategies, hiring the appropriate team and signing a lease at a great location. 

Note 1: For the simplicity of this project we are mainly focusing on the latter aspect. 

Accordingly, this project aims to use location database tool, namely Foursquare API and other datasets to choose a location in Toronto that supports both internal and external analysis conducted and maximizes the chances to successfully grow this new venture.  

Note 2: We will use data to support the following

### Internal Analysis Aspects

* Financial 
* Marketing 
* Operations 
* Human Resources 

### External Analysis Aspects
