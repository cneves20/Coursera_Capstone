# Capstone Project: Finding an Ideal Location to Open an Art Gallery in Downtown Toronto

## 1. Introduction

### 1.1. Background

The following project is a part of IBM's Data Science Professional Certificate specialization provided by Coursera. In general, this project aims to provide learners with a great insight into the various methodologies, skills and tools used by data scientists to deliver data-driven solutions and solve real-world problems. Accordingly, a scenario in which an entrepreneur is considering investing in a business idea for an art gallery venue in the city of Toronto (downtown) was set up.

At the outset, there are numerous factors to take into consideration to evaluate this business idea such as determining product and services catalogue, delineating business structure and marketing strategies, gathering the appropriate team and signing a lease at a great location. For the sake of simplicity, this project mainly focuses on the latter aspect. In other words, it aims to use machine learning and location database tools, namely Foursquare API and other datasets to choose a location in Toronto that supports both internal and external analysis needed to maximize the chances to successfully grow this new venture.


### 1.2. Problem

One crucial aspect of starting a business is to decide where to locate it. For most businesses, location plays a huge role in attracting and retaining the workforce, financing real estate, managing operating costs, and reaching the target audience. These factors also hold for an art gallery. For instance, it is ideal to centre the business where it can draw young curious and creative minds and real estate is cost-effective. The location should also promote competitive advantage by minimizing the cost to produce goods while inversely maximizing profit. Furthermore, location also impacts accessibility in terms of market, customers, resources and transportation. Therefore, the following question emerges, what is an ideal location to open an art gallery in Toronto?

## 2. Data Acquisition & Cleaning

### 2.1. Data Source

All data on postal code, boroughs and neighbourhoods in Toronto were retrieved from Wikipedia:https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M and converted into a data frame for further analysis. The data on geographical coordinates was retrieved from: http://cocl.us/Geospatial_data_ and also converted into a data frame for further analysis. Finally, the data on venues in the city of Toronto were collected via Foursquare’s API.

### 2.2. Data Cleaning 

Initially, the dataset was scraped from Wikipedia to create the data frame below.



As illustrated, the data frame consists of three columns, namely postal code, borough, and neighbourhood. There is a fair amount of missing values in the dataset, which can reduce the statistical power of the investigation and can produce biased estimates, leading to invalid conclusions. Therefore, there will be an omission of rows that do not contain an assigned value for the borough column considering that there will still be a large enough sample to process the data. Moreover, it is known that more than one neighbourhood can exist in one postal code area. For instance, the table on the Wikipedia page may have a postal code listed twice with different neighbourhoods. To avoid data duplication, it is ideal to combine rows with the same postal code into one row with the neighbourhoods separated with a comma. Also, if a row has a borough but a not assigned neighbourhood, then the neighbourhood will be the same as the borough.

To utilize the Foursquare location data, it is necessary to get the latitude and the longitude coordinates of each neighbourhood. That can be achieved by merging the geographical coordinates dataset with the data frame created.

There are many aspects to consider to select a location that will impact the growth of the business such as proximity and accessibility to people, as well as proximity to other businesses. Many believe there are advantages to clustering similar businesses in a specific area.
